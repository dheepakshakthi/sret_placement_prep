## creating a class

```python
class Dog:
    def __init__(self, name): #this method is called whenever an object is created from the class
        self.name = name #instance variable can be used in this class using self
        print("Dog object created with name:", self.name)

    def bark(self):
        return "Woof!"

d = Dog("Buddy") #d is an object of class Dog
print(type(Dog))
print(type(d)) # this is executed in the main module thats why its returning the type as __main__.Dog
print(d.bark())
```

## example of basic class implementation

```python
class students:
    def __init__(self, name, age, grade):
        self.name = name
        self.grade = grade
        self.age = age
    def getGrade(self):
        return self.grade

class courses:
    def __init__(self, course_name, maxStudents):
        self.course_name = course_name
        self.maxStudents = maxStudents
        self.studentsEnrolled = []

    def addStudent(self, studentName):
        if len(self.studentsEnrolled) < self.maxStudents:
            self.studentsEnrolled.append(studentName)
            print(f"Student {studentName} added to the course {self.course_name}.")
        else:
            print(f"Cannot add {studentName}. Course {self.course_name} is full.")
    def removeStudent(self, studentName):
        if studentName in self.studentsEnrolled:
            self.studentsEnrolled.remove(studentName)
            print(f"Student {studentName} removed from the course {self.course_name}.")
        else:
            print(f"Student {studentName} is not enrolled in the course {self.course_name}.")
    def getAverageGrade(self):
        value = 0
        for student in self.studentsEnrolled:
            value = value + student.getGrade()
        return value / len(self.studentsEnrolled)

s1 = students("Alice", 20, 85)
s2 = students("Bob", 22, 90)
s3 = students("Charlie", 19, 78)
course = courses("Math 101", 2)
course.addStudent(s1)   
course.addStudent(s2)
course.addStudent(s3)  # This should indicate that the course is full
print("Average Grade:", course.getAverageGrade())
```

## inheritence concept

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def speak(self):
        return "animal sound not defined"
    
class cat(Animal):
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color
    def speak(self):
        return "Meow!"
```

## class attributes

```python
class Person:
    class_attribute = 2

    def __init__(self, name):
        self.name = name

p1 = Person("John")
p2 = Person("Doe")
print(p1.class_attribute)  # Accessing class attribute via instance
print(Person.class_attribute)  # Accessing class attribute via class
p1.class_attribute = 5  # Modifying class attribute via instance
print(p1.class_attribute)  # Now this will print 5
print(p2.class_attribute)  # This will still print 2
Person.class_attribute = 10  # Modifying class attribute via class
print(p1.class_attribute, p2.class_attribute, Person.class_attribute)  # p1 will print 5, p2 will print 10
```

```python
class Person:
    classAttribute = 0
    def __init__(self, name):
        self.name = name
        Person.classAttribute += 1  # Incrementing class attribute

p1 = Person("John")
p2 = Person("Doe")
print(Person.classAttribute)  # This will print 2 because two instances have been created
```

## class methods
- use of decorators

```python
class Person:
    numberOfPeople = 0
    def __init__(self, name):
        self.name = name
        Person.addPerson()
    
    @classmethod #define a method that works with the class itself rather than instance
    def addPerson(cls): #cls refers to the class itself
        cls.numberOfPeople += 1
    
    @classmethod
    def getNumberOfPeople(cls):
        return cls.numberOfPeople

p1 = Person("Alice")
p2 = Person("Bob")
print(Person.getNumberOfPeople())  # This will print 2
print(p1.getNumberOfPeople())  # This will also print 2
```

## static methods

they dont need an instance of the class or act on behalf of the class itself. they are logical functions structured within the class.

```python
class Mathematics:
    @staticmethod
    def add(a, b):
        return a + b

print(Mathematics.add(5, 7))  # This will print 12
```

## multi-level inheritence

```python
class Grandparent:
    def fun1(self):
        print("I am the Grandparent.")

class Parent(Grandparent):
    def fun2(self):
        print("I am the Parent.")

class Child(Parent):
    def fun3(self):
        print("I am the Child.")

# Create an object of the Child class
obj = Child()

# Call methods from all levels of the hierarchy
obj.fun1()
obj.fun2()
obj.fun3()
```

## Abstraction

```python
from abc import ABC, abstractmethod

class demo(ABC):

    @abstractmethod
    def display1(self):
        print("abstract method")

    def display2(self):
        print("concrete method")

class child(demo):
    def display2(self):
        print("implementation of abstract method")

o = child()  # This will raise an error because we cannot instantiate an abstract class
o.display1()  # This will also raise an error
o.display2()  # This will work if the above lines are commented out
```
